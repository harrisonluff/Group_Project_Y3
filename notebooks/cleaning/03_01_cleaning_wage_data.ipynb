```python
import numpy as np
import pandas as pd
```

**A - Average Monthly Gross Wages and Salaries - P2497:**

```python
wage = pd.read_csv(repo_root / "raw/03_01_wage_data/03_01_p2497_avg_month_gross_wage.csv", delimiter=";")
wage.head()
```

```python
df = wage.copy()

# Drop unnamed column
df.drop("Unnamed: 26", axis=1, inplace=True)

# Melt to long format
id_cols = ["Code", "Name"]
long = df.melt(
    id_vars=id_cols, 
    var_name="measure",
    value_name="value"
)

# Split measure into parts
parts = long["measure"].str.split(";", expand=True)
long["type"] = parts[0].str.strip()
long["year"] = pd.to_numeric(parts[1]).astype("Int64")

out = long.rename(
        columns={"Code": "code", "Name": "powiat"}
    ).loc[
        :, ["code", "powiat", "type", "year", "value"]
    ]

out.to_csv(repo_root / "cleaned/03_01_outcome_data/wage_powiat_p2497.csv")
```
